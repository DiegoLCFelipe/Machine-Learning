```python
import pandas as pd
import os
```

```python
PATH = os.getcwd()
dados_cultura_area = pd.read_excel(PATH + '\\data\\tabela9026.xlsx',
                      skiprows=1)
dados_cultura_area.head()
```

```python
dados_cultura_area.drop(['Variável',
            'Grande Região',
            'Condição do produtor em relação às terras',
            'Grupos de atividade econômica',
            'Cor ou raça do produtor',
            'Ano'],
            axis=1,
            inplace=True)

dados_cultura_area.rename(columns={'Unnamed: 7':'Área (Ha)'}, inplace=True)
dados_cultura_area = dados_cultura_area[dados_cultura_area['Área (Ha)']!='X']
dados_cultura_area.dropna(inplace=True)
dados_cultura_area['Área (Ha)'] = dados_cultura_area['Área (Ha)'].astype(int)
dados_cultura_area.sort_values(by='Área (Ha)', ascending=False, inplace=True)
dados_cultura_area.head()
```
