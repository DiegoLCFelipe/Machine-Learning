```python
import pandas as pd
```

```python
PATH = r'data\estudo_producao.xlsx'
VAR_METRICAS = ['AnoPlantio', 'AnteseHa', 'FecundaHa', 'CachoVerdeHa', 'CachoMaduroHa', 'ProducaoTon']
dados = pd.read_excel(PATH, decimal=',')
dados.drop_duplicates(subset=['Parcela','DataColheita'], inplace=True)
dados = dados[dados['DataColheita']>='2020-03-15']
#dados.sort_values(by=['Parcela', 'DataColheita'], inplace=True)
#dados_test = dados.groupby('Parcela').apply(lambda x: x.set_index('DataColheita').asfreq('M', fill_value=None))
```

```python
dados_analise = dados[['AnoPlantio', 'DataColheita','AnteseHa', 'FecundaHa', 'CachoVerdeHa', 'CachoMaduroHa', 'ProducaoTon']].groupby(by='DataColheita').mean()
dados_analise
```

```python
dados_analise.index
```

```python

dados_analise['AnteseHa_1'] = dados_analise['AnteseHa'].shift(1)
dados_analise['AnteseHa_2'] = dados_analise['AnteseHa'].shift(2)
dados_analise['AnteseHa_3'] = dados_analise['AnteseHa'].shift(3)
dados_analise['AnteseHa_4'] = dados_analise['AnteseHa'].shift(4)
dados_analise['AnteseHa_5'] = dados_analise['AnteseHa'].shift(5)
dados_analise['AnteseHa_6'] = dados_analise['AnteseHa'].shift(6)
dados_analise['CachoVerdeHa_1'] = dados_analise['CachoVerdeHa'].shift(1)
dados_analise['CachoMaduroHa_1'] = dados_analise['CachoMaduroHa'].shift(1)
dados_analise.dropna(inplace=True)
dados_analise.sort_index()
```
