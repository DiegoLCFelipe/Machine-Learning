```python
import pandas as pd
```

```python
PATH = r'data\estudo_producao.xlsx'
VAR_METRICAS = ['AnoPlantio', 'AnteseHa', 'FecundaHa', 'CachoVerdeHa', 'CachoMaduroHa', 'ProducaoTonHa']
dados = pd.read_excel(PATH, decimal=',')
dados = dados[dados['AnoColheita']==2023]
dados = dados.set_index('DataColheita')
dados.head()
```

```python
dados_analise = dados[VAR_METRICAS]
dados_analise['AnteseHa_4'] = dados_analise['AnteseHa'].shift(120)
dados_analise['AnteseHa_5'] = dados_analise['AnteseHa'].shift(150)
dados_analise['AnteseHa_6'] = dados_analise['AnteseHa'].shift(180)
dados_analise.dropna(inplace=True)
dados_analise
```

```python
dados_analise.to_excel(r'data\teste.xlsx')
```
